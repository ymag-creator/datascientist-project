```python
import pandas as pd

#Incidents

# Ajuster les paramètres pour afficher toutes les lignes et colonnes
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

df1 = pd.read_csv(
    "../data/LFB Incident data from 2009 - 2017.csv", sep=",", low_memory=False
)
df2 = pd.read_csv(
    "../data/LFB Incident data from 2018 onwards.csv", sep=",", low_memory=False
)

df_incidents = pd.concat([df1, df2], axis=0)
df_incidents.reset_index(drop=True)

print(df_incidents.shape)
display(df_incidents.head())

# suppression des .00
df_incidents["IncidentNumber"] = df_incidents["IncidentNumber"].apply(
    lambda x: str(x)[0:-3] if (".00" in x) & (str(x)[-3:] == ".00") else x
)
display(df_incidents.head())
display(df_incidents.info())
display(df_incidents.describe())
```

```python
# Mobilisations

df1 = pd.read_csv(
    "../data/LFB Mobilisation data from January 2009 - 2014.csv",
    sep=";",
    low_memory=False,
)
df2 = pd.read_csv(
    "../data/LFB Mobilisation data from 2015 - 2020.csv", sep=";", low_memory=False
)
df3 = pd.read_csv(
    "../data/LFB Mobilisation data from 2021 - 2024.csv",
    sep=",",
    low_memory=False,
    usecols=lambda column: column not in ["BoroughName", "WardName"],
)

df_mobilisations = pd.concat([df1, df2, df3], axis=0)
df_mobilisations.reset_index(drop=True)

display(df_mobilisations.head())
display(df_mobilisations.info())
display(df_mobilisations.describe())
```

```python
# Conversion des dates

# DateOfCall 09-Aug-21
# TimeOfCall 00:04:25
# DateAndTimeMobilised 01/01/2018 00:04
# DateAndTimeMobile 01/01/2018 00:04
# DateAndTimeArrived 01/01/2018 00:04
# DateAndTimeLeft 01/01/2018 00:16
# DateAndTimeReturned 01/01/2018 00:16

# pd.to_datetime('2030-26-10', format='%Y-%d-%m')
# pd.to_datetime('2018-10-26 12:00:00', format='%Y-%m-%d %H:%M:%S')


df_incidents.DateOfCall = pd.to_datetime(df_incidents.DateOfCall, format="%d-%b-%y")
df_incidents.TimeOfCall = pd.to_datetime(df_incidents.TimeOfCall, format="%H:%M:%S")
df_mobilisations.DateAndTimeMobilised = pd.to_datetime(
    df_mobilisations.DateAndTimeMobilised, format="%d/%m/%Y %H:%M"
)
df_mobilisations.DateAndTimeMobile = pd.to_datetime(df_mobilisations.DateAndTimeMobile, format="%d/%m/%Y %H:%M")
df_mobilisations.DateAndTimeArrived = pd.to_datetime(
    df_mobilisations.DateAndTimeArrived, format="%d/%m/%Y %H:%M"
)
df_mobilisations.DateAndTimeLeft = pd.to_datetime(
    df_mobilisations.DateAndTimeLeft, format="%d/%m/%Y %H:%M"
)
df_mobilisations.DateAndTimeReturned = pd.to_datetime(
    df_mobilisations.DateAndTimeReturned, format="%d/%m/%Y %H:%M"
)
```

```python
# mise de mobilisations sur une ligne

# ajout d'une colonne PumpSecondsOnSite
df_mobilisations["PumpOnSiteSeconds"] = (df_mobilisations.DateAndTimeLeft - df_mobilisations.DateAndTimeArrived).dt.seconds
print(df_mobilisations.shape)

display(df_mobilisations.head(10))

# création des temps min/max par incidentNumber
aggregated = (
    df_mobilisations.groupby("IncidentNumber")  # .head(100)
    .agg(
        PumpSecondsOnSite_min=("PumpOnSiteSeconds", "min"),
        PumpSecondsOnSite_max=("PumpOnSiteSeconds", "max"),
        TurnoutTimeSeconds_min=("TurnoutTimeSeconds", "min"),
        TurnoutTimeSeconds_max=("TurnoutTimeSeconds", "max"),
        TravelTimeSeconds_min=("TravelTimeSeconds", "min"),
        TravelTimeSeconds_max=("TravelTimeSeconds", "max"),
    )
    .reset_index()
)
display(aggregated.head(10))

# jointure sur df_mobilisations pour avoir les min/max
merge = pd.merge(
    df_mobilisations, aggregated, left_on="IncidentNumber", right_on="IncidentNumber", how="left"
)
display(merge.head(10))
print("merge.shape", merge.shape)

print("merge.IncidentNumber.duplicated().sum()", merge.IncidentNumber.duplicated().sum())

# suppression des doublons pour garder la 1ere ligne
df_mobilisations = merge.drop_duplicates(subset=["IncidentNumber"], keep="first")
print("df_mobilisations.shape", df_mobilisations.shape)
display(df_mobilisations.head(10))
```

```python
print(df_incidents.shape)
print(df_mobilisations.shape)

# Merge
df = pd.merge(

    df_incidents,
    df_mobilisations,
    left_on="IncidentNumber",
    right_on="IncidentNumber",
    how="left",
)


print(df.shape)
display(df.head())
display(df.info())
display(df.describe())
```
